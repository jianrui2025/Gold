```python
import json
import adata
import statistics
```

```python
df = adata.fund.info.all_etf_exchange_traded_info()
df = df.to_dict(orient="records")
log = {}
for fund_code in df:
    code = fund_code["fund_code"]
    try:
        tmp = adata.fund.market.get_market_etf(fund_code=code)
    except:
        print(fund_code)
        continue
    tmp = tmp.to_dict(orient='records')[-21:-1]
    log[code] = statistics.mean([int(i["volume"]) for i in tmp])
```

```python
log_list = [(k,v) for k,v in log.items()]
log_list.sort(key=lambda x: x[1])
log_list_select = [i[0] for i in log_list if i[1]>60000000.0]
```
