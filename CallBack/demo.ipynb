```python
import json
import os
import copy

path_all = "/home/jianrui/workspace/Gold/CallBack/log_MeanLineAndVolume/"

def get_code(path1):
    files1 = [os.path.join(path1, f) for f in os.listdir(path1) if os.path.isfile(os.path.join(path1, f))]
    files1 = set([i.split("/")[-1] for i in files1])
    return files1

def precise(i):
    log = i["log"]
    buy_num = 0
    yingli_num = 0
    for index,k in enumerate(log):
        if index%2 == 0:
            buy_num += 1
        else:
            if "盈利卖出" in log[k]:
                yingli_num += 1
    if buy_num != 0:
        i["precion"] = yingli_num/buy_num
    else:
        i["precion"] = 0

def read_dict(path1,fund_code):
    for code in fund_code:
        with open(path1+code,"r") as f:
            re1 = [json.loads(i.strip()) for i in f]
            re1 = [i for i in re1 if i["mean_long_day"]>i["mean_short_day"]]
            
            re1_dict = {"-".join([str(i) for i in [i["mean_long_day"],i["mean_short_day"],i["Volume_day"]]]):i for i in re1}
            re1_key = set([i for i in re1_dict])
        
        for i in re1:
            precise(i)
        
        

fund_code = get_code(path_all)
read_dict(path_all,fund_code)
```

```python
import json
import os
import copy

path = "/home/jianrui/workspace/Gold/CallBack/conf/MeanLineAndVolume_v8.jsonl"

def read_path(path):
    with open(path,"r") as f:
        re = [json.loads(i.strip()) for i in f]
    return re

def get_day(re):
    # 按照日期，记录股票买入时间点
    re_day = {}
    for i in re:
        log = i["log"]
        for day,v in log.items():
            if "买入" in v:
                re_day.setdefault(day,[])
                re_day[day].append(i["fund_code"])
    keys = list(re_day.keys())
    keys.sort(key=lambda x: int(x))
    for i in keys:
        print(i, " ".join(re_day[i]))
    print(len(keys))

re = read_path(path)
get_day(re)
```
